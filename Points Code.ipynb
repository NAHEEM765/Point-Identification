## PhantomMain Function

```python
# Inputs: coordinates, main diagonals, testM, tM, rem, testm, minor diagonals, tm, 
def phantomMain(coordinates, M, m):
    # Original Coordinates sorting with Main Diagonal Coordinates
    testM = [] 
    for i in range(len(coordinates)):
        for j in range(len(M)):
            if coordinates[i] in M[j]:
                testM += (M[j])
    tM = list(dict.fromkeys(testM))
    print(f"The coordinates we are working with are {coordinates}")
    print(f"These are the coordinates of the main diagonals {M}")
    # print(f"len(tM), tM)" 
    print(f"tM: {tM}")
    rem_m1 = [l for l in tM if l not in coordinates] # Number associated with m=1
    rem = [l for l in tM if l not in coordinates] # Remainder to be used after
    print(f"These are the phantom points after adding slope lines of m=1: {rem_m1}")

    # Remainder sorting with Minor Diagonal Coordinates
    testm = [] 
    for i in range(len(coordinates)):
        for j in range(len(m)):
            if coordinates[i] in m[j]:
                testm += (m[j])
    tm = list(dict.fromkeys(testm))
    print(f"tm: {tm}")

    
    for i in tM:
        if i not in tm:
            rem.remove(i)
    print(f"These are the coordinates of the minor diagonal {m}")     
    if len(rem) == 0:
        print(f"The are no remaining phantom points after adding slope lines for m=-1")
    elif len(rem) == 1:
        print(f"The remaining phantom point after adding slope lines for m=-1 is: {rem}"
              f"\nthere is only {len(rem)} phantom point")
    else:
        print(f"The remaining phantom points after adding slope lines for m=-1 are: {rem}"
              f"\nthere are {len(rem)} phantom points")
    return rem_m1, rem

# ToDo:
# Customize function so that, if you are dealing with one, two or three points, the function recognizes that and will 
# compare the list input according to that length
# Use function and configurations to check for the slopeof 1
# All but 2 configurations (Case 9 - #14 and #21) don't have zero phantom points after adding slope lines for m=-1
```

## Configuration Reveal Functions

```python
def dummy(coordinates, M, m):
    # Original Coordinates sorting with Main Diagonal Coordinates
    testM = [] 
    for i in range(len(coordinates)):
        for j in range(len(M)):
            if coordinates[i] in M[j]:
                testM += (M[j])
    tM = list(dict.fromkeys(testM))
    # print(f"len(tM), tM)" 
    rem_m1 = [l for l in tM if l not in coordinates] # Number associated with m=1
    rem = [l for l in tM if l not in coordinates] # Remainder to be used after

    # Remainder sorting with Minor Diagonal Coordinates
    testm = [] 
    for i in range(len(coordinates)):
        for j in range(len(m)):
            if coordinates[i] in m[j]:
                testm += (m[j])
    tm = list(dict.fromkeys(testm))

    
    for i in tM:
        if i not in tm:
            rem.remove(i)
    return rem_m1, rem

def func(coordinates, num, M, m):          # Takes as input the coordinates, the number of elements in each combination, the main diagonal coordinates as well as the minor diagonal coordinates
    for i in range(num+1):
        a = 0
        b = 0
        L = com(coordinates, i)        # Creates the combinations
        ind = []
        for j in L:                      # Iterates over each combination
            m_1, m_n1 = dummy(j, M, m)   # get the number of phantom points after adding slope lines of 1 and -1 respectively
            if len(m_1) == 0:
                a+=1
            if len(m_1) != 0 and len(m_n1) == 0:
                b+=1
                ind.append(L.index(j))
        print(f"\nThese are the indices where after adding slope lines of m=-1, the number of phantom points is 0: {ind}")        
        if a != 1:
            print(f"For configurations with {i} points, there are {a} configurations revealed after adding slope lines of m=1, and \n{b-a} configurations revealed after adding slope lines of m=-1")
        if a == 1:
            print(f"For configurations with {i} points, there is {a} configuration revealed after adding slope lines of m=1, and \n{b-a} configurations revealed after adding slope lines of m=-1")
    return a, (b-a)
```

```python
def funcdum(coordinates, num, M, m):        # Takes as input the coordinates and the number of elements in each combination
        a = 0
        b = 0
        L = com(coordinates, num)        # Creates the combinations                    # Iterates over each combination
        for i in L:                      # Iterates over each combination
            m_1, m_n1 = dummy(i, M, m)   # get the number of phantom points after adding slope lines of 1 and -1 respectively
            if len(m_1) == 0:
                a+=1
            if len(m_n1) == 0:
                b+=1
        # if a != 1:
        #     print(f"For configurations with {i} points, there are {a} configurations revealed after adding slope lines of m=1, and \n{b-a} configurations revealed after adding slope lines of m=-1.")
        # if a == 1:
        #     print(f"For configurations with {i} points, there is {a} configuration revealed after adding slope lines of m=1, and \n{b-a} configurations revealed after adding slope lines of m=-1.")
        return a, a, b

def table(coordinates, num, M, m):
    NPC = []
    headers = ["No.", "m1","mn1", "m1mn1"] # NC0 - Number of Known Points, NCb0 - Number of Revealed configurations before adding slope lines of m=-1, NCA0 - Number of Revealed configurations after adding slope lines of m=-1
    for i in range(num+1):
        a, b, c = funcdum(coordinates, i, M, m)
        NPC.append([i, a, b, (c-a)])
    print(tabulate(NPC, headers=headers, tablefmt="grid"))
```
